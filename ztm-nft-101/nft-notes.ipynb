# Non-Fungible Tokens (NFT)s

- smart contract - create tokens
- tokens are managed as they transfer from one to another
- create an NFT = **mint**
---
- fungible = item can be replaced by another identical item without any change in behaviour 
- coins are fungible, 1 GBP coin is much the same as any other 1 GBP
- bitcoins, ether, etc = fungible
- a NFT token is unique
- any art work - wyou can have copies/imitations = but the original is non-fungible 
- fungility = subjective
- how valuable is it to you

- we use smart contracts to not only create NFT but to manage it

## What's the point of NFTs

- allow us to represent Non-fungible items digitally as tokens
- can be used to prove ownership of items on the internet 
- we can then buy/sell/transfer it 
---
- internet is full of digital things
- nearly impossible to show that you own a digital thing 
- imagine if you created a viral video
- how do you control the narative - how do you prove you had the original 
---
- if you'd uploaded it to an IPFS, and minted a token, then it's easy to prove you were the creator 
- proven using a trail of transactions on blockchain 
- unfortunately - you can also upload something to nftTube before the original owner did 
- then you are seen as the owner
- needs good info from the start 
---
NFT = not the item - but who owns the item!
